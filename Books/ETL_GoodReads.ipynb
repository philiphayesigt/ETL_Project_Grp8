```python
#Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#import csv into a datframe
csv_file = "books.csv"
# books_df = pd.read_csv(csv_file)
books_df = pd.read_csv(csv_file, error_bad_lines=False,warn_bad_lines=False)
books_df.head()
```

```python
#Create new dataframe with select columns
new_books_df = books_df[['bookID','title','authors','average_rating','isbn','language_code','ratings_count']]
new_books_df.head()
```

```python
# Placing the bookid's in a list
tolst_books = list(new_books_df['bookID'])

new_lst_books = []

#Appending the book id with a 'B' infront
for i in enumerate(tolst_books):
     new_lst_books.append('B'+str(i[1]))

#Adding the modified bookid to a dataframe along with the other columns.
new_books_df1 = pd.DataFrame(new_books_df)

new_books_df1['bookid'] = new_lst_books  
new_books_df1
```

```python
#Dropping the old bookid column.
clean_book_df = new_books_df1.drop('bookID', axis=1)
clean_book_df.head()
```

```python
#Connect to local database
rds_connection_string = "postgres:vasu12@localhost:5432/ETL_Grp_8"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
#change column names to match db tables columns
clean_book_db = clean_book_df.rename(columns={'average_rating':'rating',
                                           'language_code':'lang_code'})

clean_book_db.head()
```

```python
#Modify data to suite the database
titles_df = new_books_df1[['bookid','title']]
titles_df.head()
```

```python
#Inserting the dataframe data into the books table.
clean_book_db.to_sql(name='books', con=engine, if_exists='append', index=False)
```

```python
clean_titles_df = titles_df.rename(columns={'bookid':'id'})
```

```python
clean_titles_df.head()
```

```python
#Inserting data into titles table (bookid and title)
clean_titles_df.to_sql(name='titles', con=engine, if_exists='append', index=False)
```

```python
#Querying books table
pd.read_sql_query('select * from books', con=engine).head()
```

```python
#Querying the titles table
pd.read_sql_query('select * from titles', con=engine).head()
```

```python
# Querying the database to retreive the books based on the keyword.
pd.read_sql_query(r"select b.bookid,b.title,b.rating from books b, titles t where b.bookid = t.id and t.title like '%%Harry%%'",con=engine)
```
